```python
import tensorflow as tf

# Convert the model
converter = tf.lite.TFLiteConverter.from_keras_model(tf.keras.models.load_model(r'E:\Pranav\Hackathons\NFC3_Lorem-Ipsum\TSECNFCBackend\TsecAPI\model\model.h5'))
converter.target_spec.supported_ops = [
    tf.lite.OpsSet.TFLITE_BUILTINS,
    tf.lite.OpsSet.SELECT_TF_OPS
]
tflite_model = converter.convert()
# Save the model
with open(r'E:\Pranav\Hackathons\NFC3_Lorem-Ipsum\TSECNFCBackend\TsecAPI\model\model.tflite', 'wb') as f:
    f.write(tflite_model)
```
